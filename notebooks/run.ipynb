```python
from typing import List
import os
import pandas as pd
import torch

from benchmarl.algorithms import MappoConfig
from benchmarl.environments import VmasTask
from benchmarl.experiment import Experiment, ExperimentConfig
from benchmarl.experiment.callback import Callback
from benchmarl.models.mlp import MlpConfig
from tensordict import TensorDictBase
```

```python
class SwarmMetricsCallback(Callback):

    def __init__(self, save_path="/content/swarm_metrics.csv"):
        super().__init__()
        self.save_path = save_path
        self.rows = []
        self.evaluation_counter = 0

    def _try_get(self, td, keys):
        for key in keys:
            try:
                value = td.get(key)
                if value is not None:
                    return value
            except Exception:
                pass
        return None

    def _mean_tensor(self, tensor):
        if tensor is None:
            return None

        if isinstance(tensor, torch.Tensor):
            return float(tensor.detach().cpu().float().mean())

        return float(tensor)

    def on_evaluation_end(self, rollouts: List[TensorDictBase]):

        print(
            f"\n========== Evaluation {self.evaluation_counter} =========="
        )

        for episode_id, rollout in enumerate(rollouts):

            reward = self._try_get(
                rollout,
                [
                    ("next", "agents", "reward"),
                    ("agents", "reward"),
                    "reward",
                ],
            )

            action = self._try_get(
                rollout,
                [
                    ("agents", "action"),
                    "action",
                ],
            )

            dist_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "dist_rew"),
                ],
            )

            rot_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "rot_rew"),
                ],
            )

            pos_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "pos_rew"),
                ],
            )

            wind_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "agent_wind_rew"),
                ],
            )

            vel_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "agent_vel_rew"),
                ],
            )

            energy_rew = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "agent_energy_rew"),
                ],
            )

            delta_vel_goal = self._try_get(
                rollout,
                [
                    ("next", "agents", "info", "delta_vel_to_goal"),
                ],
            )

            control_effort = None
            control_smoothness = None

            if action is not None:

                action = action.detach().cpu()

                control_effort = float(
                    (action ** 2).sum(dim=-1).mean()
                )

                if action.shape[0] > 1:

                    da = action[1:] - action[:-1]

                    control_smoothness = float(
                        (da ** 2).sum(dim=-1).mean()
                    )

            row = {
                "evaluation": self.evaluation_counter,
                "episode": episode_id,

                "reward_mean":
                    self._mean_tensor(reward),

                "dist_rew":
                    self._mean_tensor(dist_rew),

                "rot_rew":
                    self._mean_tensor(rot_rew),

                "pos_rew":
                    self._mean_tensor(pos_rew),

                "agent_wind_rew":
                    self._mean_tensor(wind_rew),

                "agent_vel_rew":
                    self._mean_tensor(vel_rew),

                "agent_energy_rew":
                    self._mean_tensor(energy_rew),

                "delta_vel_to_goal":
                    self._mean_tensor(delta_vel_goal),

                "control_effort":
                    control_effort,

                "control_smoothness":
                    control_smoothness,
            }

            self.rows.append(row)

        pd.DataFrame(self.rows).to_csv(
            self.save_path,
            index=False
        )

        print(
            f"Métricas salvas em {self.save_path}"
        )

        self.evaluation_counter += 1
```

```python
class UniversalSwarmMetricsCallback(Callback):

    def __init__(
        self,
        save_path="/content/universal_swarm_metrics.csv",
        collision_threshold=0.08,
        print_keys=True,
    ):
        super().__init__()
        self.save_path = save_path
        self.collision_threshold = collision_threshold
        self.print_keys = print_keys
        self.rows = []
        self.evaluation_counter = 0

    def _safe_get(self, td, key):
        try:
            value = td.get(key)
            if value is not None:
                return value
        except Exception:
            return None
        return None

    def _mean(self, x):
        if x is None:
            return None
        if isinstance(x, torch.Tensor):
            if x.numel() == 0:
                return None
            return float(x.detach().cpu().float().mean())
        try:
            return float(x)
        except Exception:
            return None

    def _std(self, x):
        if x is None:
            return None
        if isinstance(x, torch.Tensor):
            if x.numel() <= 1:
                return 0.0
            return float(x.detach().cpu().float().std())
        return None

    def _min(self, x):
        if x is None:
            return None
        if isinstance(x, torch.Tensor):
            if x.numel() == 0:
                return None
            return float(x.detach().cpu().float().min())
        return None

    def _max(self, x):
        if x is None:
            return None
        if isinstance(x, torch.Tensor):
            if x.numel() == 0:
                return None
            return float(x.detach().cpu().float().max())
        return None

    def _find_first_available(self, rollout, keys):
        for key in keys:
            value = self._safe_get(rollout, key)
            if value is not None:
                return value, key
        return None, None

    def _compute_action_metrics(self, action):
        metrics = {
            "control_effort": None,
            "control_smoothness": None,
            "action_mean": None,
            "action_std": None,
            "action_min": None,
            "action_max": None,
        }

        if action is None:
            return metrics

        action = action.detach().cpu().float()

        metrics["action_mean"] = float(action.mean())
        metrics["action_std"] = float(action.std())
        metrics["action_min"] = float(action.min())
        metrics["action_max"] = float(action.max())

        metrics["control_effort"] = float(
            (action ** 2).sum(dim=-1).mean()
        )

        if action.shape[0] > 1:
            delta_action = action[1:] - action[:-1]
            metrics["control_smoothness"] = float(
                (delta_action ** 2).sum(dim=-1).mean()
            )

        return metrics

    def _compute_swarm_metrics(self, loc):
        metrics = {
            "position_source_found": False,
            "swarm_dispersion": None,
            "mean_interagent_distance": None,
            "min_interagent_distance": None,
            "max_interagent_distance": None,
            "collision_rate": None,
            "swarm_diameter": None,
        }

        if loc is None:
            return metrics

        loc = loc.detach().cpu().float()

        if loc.shape[-1] < 2:
            return metrics

        loc = loc[..., :2]

        n_agents = loc.shape[-2]

        if n_agents < 2:
            return metrics

        metrics["position_source_found"] = True

        centroid = loc.mean(dim=-2, keepdim=True)

        dispersion = ((loc - centroid) ** 2).sum(dim=-1).mean()

        pairwise_dist = torch.cdist(loc, loc)

        mask = ~torch.eye(
            n_agents,
            dtype=torch.bool,
            device=pairwise_dist.device
        )

        valid_dist = pairwise_dist[..., mask]

        metrics["swarm_dispersion"] = float(dispersion)
        metrics["mean_interagent_distance"] = float(valid_dist.mean())
        metrics["min_interagent_distance"] = float(valid_dist.min())
        metrics["max_interagent_distance"] = float(valid_dist.max())
        metrics["swarm_diameter"] = float(valid_dist.max())

        metrics["collision_rate"] = float(
            (valid_dist < self.collision_threshold).float().mean()
        )

        return metrics

    def _extract_all_numeric_info(self, rollout):
        """
        Extrai automaticamente qualquer chave numérica com 'info'.
        Exemplo:
        ('next', 'agents', 'info', 'dist_rew')
        vira:
        info_next_agents_dist_rew_mean
        """

        metrics = {}

        try:
            keys = list(rollout.keys(True, True))
        except Exception:
            return metrics

        for key in keys:

            if not isinstance(key, tuple):
                continue

            if "info" not in key:
                continue

            value = self._safe_get(rollout, key)

            if value is None:
                continue

            if not isinstance(value, torch.Tensor):
                continue

            if not torch.is_floating_point(value) and not torch.is_tensor(value):
                continue

            name = "info_" + "_".join(str(k) for k in key)

            metrics[f"{name}_mean"] = self._mean(value)
            metrics[f"{name}_std"] = self._std(value)
            metrics[f"{name}_min"] = self._min(value)
            metrics[f"{name}_max"] = self._max(value)

        return metrics

    def _extract_observation_stats(self, observation):
        metrics = {
            "observation_mean": None,
            "observation_std": None,
            "observation_min": None,
            "observation_max": None,
        }

        if observation is None:
            return metrics

        observation = observation.detach().cpu().float()

        metrics["observation_mean"] = float(observation.mean())
        metrics["observation_std"] = float(observation.std())
        metrics["observation_min"] = float(observation.min())
        metrics["observation_max"] = float(observation.max())

        return metrics

    def on_evaluation_end(self, rollouts: List[TensorDictBase]):

        print(f"\n========== Evaluation {self.evaluation_counter} ==========")

        for episode_id, rollout in enumerate(rollouts):

            if self.print_keys and self.evaluation_counter == 0 and episode_id == 0:
                print("\nChaves disponíveis no rollout:")
                for key in rollout.keys(True, True):
                    print(key)

            reward, reward_key = self._find_first_available(
                rollout,
                [
                    ("next", "agents", "reward"),
                    ("agents", "reward"),
                    ("next", "reward"),
                    "reward",
                ],
            )

            action, action_key = self._find_first_available(
                rollout,
                [
                    ("agents", "action"),
                    ("next", "agents", "action"),
                    "action",
                    ("action",),
                ],
            )

            done, done_key = self._find_first_available(
                rollout,
                [
                    ("next", "done"),
                    "done",
                    ("next", "terminated"),
                    "terminated",
                ],
            )

            observation, obs_key = self._find_first_available(
                rollout,
                [
                    ("agents", "observation"),
                    ("next", "agents", "observation"),
                    "observation",
                    ("next", "observation"),
                ],
            )

            loc, loc_key = self._find_first_available(
                rollout,
                [
                    ("agents", "loc"),
                    ("next", "agents", "loc"),
                    ("agents", "pos"),
                    ("next", "agents", "pos"),
                    ("agents", "position"),
                    ("next", "agents", "position"),
                    ("agents", "state", "pos"),
                    ("next", "agents", "state", "pos"),
                ],
            )

            row = {
                "evaluation": self.evaluation_counter,
                "episode": episode_id,

                "reward_key": str(reward_key),
                "action_key": str(action_key),
                "done_key": str(done_key),
                "observation_key": str(obs_key),
                "position_key": str(loc_key),

                "reward_mean": self._mean(reward),
                "reward_std": self._std(reward),
                "reward_min": self._min(reward),
                "reward_max": self._max(reward),

                "done_rate": self._mean(done),
            }

            row.update(
                self._compute_action_metrics(action)
            )

            row.update(
                self._compute_swarm_metrics(loc)
            )

            row.update(
                self._extract_observation_stats(observation)
            )

            row.update(
                self._extract_all_numeric_info(rollout)
            )

            self.rows.append(row)

        pd.DataFrame(self.rows).to_csv(
            self.save_path,
            index=False
        )

        print(f"Métricas salvas em: {self.save_path}")

        self.evaluation_counter += 1
```

```python
class PaperMetricsCallback(Callback):
    def __init__(
        self,
        save_path="/content/paper_metrics.csv",
        r_safe=0.40,
        r_near=0.80,
        v_max=2.0,
        a_max=1.5,
        dt=0.1,
        formation_type="circle",  # "circle", "line", "v"
        formation_spacing=1.0,
        success_formation_threshold=0.50,
        success_goal_threshold=1.0,
        print_keys=True,
    ):
        super().__init__()
        self.save_path = save_path
        self.r_safe = r_safe
        self.r_near = r_near
        self.v_max = v_max
        self.a_max = a_max
        self.dt = dt
        self.formation_type = formation_type
        self.formation_spacing = formation_spacing
        self.success_formation_threshold = success_formation_threshold
        self.success_goal_threshold = success_goal_threshold
        self.print_keys = print_keys

        self.rows = []
        self.evaluation_counter = 0

    def _safe_get(self, td, key):
        try:
            value = td.get(key)
            if value is not None:
                return value
        except Exception:
            pass
        return None

    def _find_first(self, rollout, keys):
        for key in keys:
            value = self._safe_get(rollout, key)
            if value is not None:
                return value, key
        return None, None

    def _mean(self, x):
        if x is None:
            return None
        if isinstance(x, torch.Tensor):
            if x.numel() == 0:
                return None
            return float(x.detach().cpu().float().mean())
        try:
            return float(x)
        except Exception:
            return None

    def _std(self, x):
        if x is None or not isinstance(x, torch.Tensor) or x.numel() <= 1:
            return None
        return float(x.detach().cpu().float().std())

    def _make_formation_offsets(self, n_agents, device):
        spacing = self.formation_spacing

        if self.formation_type == "line":
            x = torch.linspace(
                -(n_agents - 1) / 2,
                (n_agents - 1) / 2,
                n_agents,
                device=device,
            ) * spacing
            y = torch.zeros(n_agents, device=device)
            offsets = torch.stack([x, y], dim=-1)

        elif self.formation_type == "v":
            offsets = []
            offsets.append(torch.tensor([0.0, 0.0], device=device))
            side = 1
            level = 1
            for _ in range(1, n_agents):
                offsets.append(
                    torch.tensor(
                        [level * spacing, side * level * spacing],
                        device=device,
                    )
                )
                side *= -1
                if side == 1:
                    level += 1
            offsets = torch.stack(offsets, dim=0)

        else:
            angles = torch.linspace(
                0,
                2 * torch.pi,
                n_agents + 1,
                device=device,
            )[:-1]
            radius = spacing
            offsets = torch.stack(
                [
                    radius * torch.cos(angles),
                    radius * torch.sin(angles),
                ],
                dim=-1,
            )

        offsets = offsets - offsets.mean(dim=0, keepdim=True)
        return offsets

    def _extract_position(self, rollout):
        loc, key = self._find_first(
            rollout,
            [
                ("agents", "loc"),
                ("next", "agents", "loc"),
                ("agents", "pos"),
                ("next", "agents", "pos"),
                ("agents", "position"),
                ("next", "agents", "position"),
            ],
        )

        if loc is None:
            return None, key

        loc = loc.detach().cpu().float()

        if loc.shape[-1] < 2:
            return None, key

        return loc[..., :2], key

    def _extract_action(self, rollout):
        action, key = self._find_first(
            rollout,
            [
                ("agents", "action"),
                ("next", "agents", "action"),
                "action",
            ],
        )

        if action is None:
            return None, key

        return action.detach().cpu().float(), key

    def _extract_reward(self, rollout):
        reward, key = self._find_first(
            rollout,
            [
                ("next", "agents", "reward"),
                ("agents", "reward"),
                ("next", "reward"),
                "reward",
            ],
        )
        return reward, key

    def _extract_velocity(self, rollout):
        vel, key = self._find_first(
            rollout,
            [
                ("agents", "vel"),
                ("next", "agents", "vel"),
                ("agents", "velocity"),
                ("next", "agents", "velocity"),
            ],
        )

        if vel is None:
            return None, key

        vel = vel.detach().cpu().float()

        if vel.shape[-1] < 2:
            return None, key

        return vel[..., :2], key

    def _compute_metrics(self, loc, action, velocity):
        metrics = {
            "formation_error": None,
            "collision_rate": None,
            "collision_episode": None,
            "near_miss_rate": None,
            "near_miss_episode": None,
            "success_rate": None,
            "energy_proxy": None,
            "control_smoothness": None,
            "velocity_violation_rate": None,
            "acceleration_violation_rate": None,
            "constraint_violations": None,
            "swarm_dispersion": None,
            "mean_interagent_distance": None,
            "min_interagent_distance": None,
            "swarm_diameter": None,
        }

        if action is not None:
            metrics["energy_proxy"] = float(
                (action ** 2).sum(dim=-1).mean() * self.dt
            )

            if action.shape[0] > 1:
                da = action[1:] - action[:-1]
                metrics["control_smoothness"] = float(
                    (da ** 2).sum(dim=-1).mean()
                )

            acc_norm = torch.linalg.norm(action, dim=-1)
            metrics["acceleration_violation_rate"] = float(
                (acc_norm > self.a_max).float().mean()
            )

        if velocity is not None:
            vel_norm = torch.linalg.norm(velocity, dim=-1)
            metrics["velocity_violation_rate"] = float(
                (vel_norm > self.v_max).float().mean()
            )

        if loc is None:
            return metrics

        n_agents = loc.shape[-2]

        if n_agents < 2:
            return metrics

        centroid = loc.mean(dim=-2, keepdim=True)

        offsets = self._make_formation_offsets(
            n_agents=n_agents,
            device=loc.device,
        )

        while offsets.dim() < loc.dim():
            offsets = offsets.unsqueeze(0)

        desired_pos = centroid + offsets

        formation_error_t = torch.linalg.norm(
            loc - desired_pos,
            dim=-1,
        ).mean(dim=-1)

        metrics["formation_error"] = float(formation_error_t.mean())

        final_window = min(int(5 / self.dt), formation_error_t.shape[0])
        final_formation_error = formation_error_t[-final_window:].mean()

        pairwise_dist = torch.cdist(loc, loc)

        mask = ~torch.eye(
            n_agents,
            dtype=torch.bool,
            device=pairwise_dist.device,
        )

        valid_dist = pairwise_dist[..., mask]

        collision_events = valid_dist < self.r_safe

        near_miss_events = (
            (valid_dist >= self.r_safe)
            & (valid_dist < self.r_near)
        )

        metrics["collision_rate"] = float(collision_events.float().mean())
        metrics["collision_episode"] = float(collision_events.any())
        metrics["near_miss_rate"] = float(near_miss_events.float().mean())
        metrics["near_miss_episode"] = float(near_miss_events.any())

        metrics["mean_interagent_distance"] = float(valid_dist.mean())
        metrics["min_interagent_distance"] = float(valid_dist.min())
        metrics["swarm_diameter"] = float(valid_dist.max())

        metrics["swarm_dispersion"] = float(
            ((loc - centroid) ** 2).sum(dim=-1).mean()
        )

        no_collision = not bool(collision_events.any())
        formation_ok = bool(
            final_formation_error < self.success_formation_threshold
        )

        # Fallback de sucesso:
        # Como nem todo cenário VMAS expõe goal/waypoint,
        # aqui sucesso = formação final adequada + ausência de colisão.
        metrics["success_rate"] = float(no_collision and formation_ok)

        v_rate = metrics["velocity_violation_rate"] or 0.0
        a_rate = metrics["acceleration_violation_rate"] or 0.0

        metrics["constraint_violations"] = (
            v_rate
            + a_rate
            + metrics["collision_rate"]
        )

        return metrics

    def _extract_all_info_metrics(self, rollout):
        metrics = {}

        try:
            keys = list(rollout.keys(True, True))
        except Exception:
            return metrics

        for key in keys:
            if not isinstance(key, tuple):
                continue

            if "info" not in key:
                continue

            value = self._safe_get(rollout, key)

            if value is None or not isinstance(value, torch.Tensor):
                continue

            if value.numel() == 0:
                continue

            name = "info_" + "_".join(str(k) for k in key)

            value = value.detach().cpu().float()

            metrics[f"{name}_mean"] = float(value.mean())

            if value.numel() > 1:
                metrics[f"{name}_std"] = float(value.std())
                metrics[f"{name}_min"] = float(value.min())
                metrics[f"{name}_max"] = float(value.max())

        return metrics

    def on_evaluation_end(self, rollouts: List[TensorDictBase]):
        print(f"\n========== Evaluation {self.evaluation_counter} ==========")

        for episode_id, rollout in enumerate(rollouts):

            if self.print_keys and self.evaluation_counter == 0 and episode_id == 0:
                print("\nChaves disponíveis no rollout:")
                for key in rollout.keys(True, True):
                    print(key)

            reward, reward_key = self._extract_reward(rollout)
            loc, loc_key = self._extract_position(rollout)
            action, action_key = self._extract_action(rollout)
            velocity, velocity_key = self._extract_velocity(rollout)

            metrics = self._compute_metrics(
                loc=loc,
                action=action,
                velocity=velocity,
            )

            row = {
                "evaluation": self.evaluation_counter,
                "episode": episode_id,

                "reward_key": str(reward_key),
                "position_key": str(loc_key),
                "action_key": str(action_key),
                "velocity_key": str(velocity_key),

                "reward_mean": self._mean(reward),
                "reward_std": self._std(reward),

                **metrics,
            }

            row.update(self._extract_all_info_metrics(rollout))

            self.rows.append(row)

        pd.DataFrame(self.rows).to_csv(
            self.save_path,
            index=False,
        )

        print(f"Métricas salvas em: {self.save_path}")

        self.evaluation_counter += 1
```
